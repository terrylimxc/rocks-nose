### Import Libraries & Read File

```python
import pandas as pd
```

```python
parsed = pd.read_csv('./data/parsed_data.csv')
info = pd.read_csv('./data/data.info')
info = info.rename(columns={"transcript_position": "position"})

df = pd.merge(parsed,info,on=['transcript_id', 'position'] ,how='left')
```

#### Mean

```python
mean_ds = df.groupby(["gene_id", "transcript_id", "position"]).mean().reset_index()
mean_ds
```

```python
nuc = df[["gene_id", "transcript_id", "position", "nucleotide"]].groupby(["gene_id", "transcript_id", "position"])["nucleotide"].unique().reset_index()
nuc.nucleotide = nuc.nucleotide.apply(lambda x: x[0])
nuc
```

```python
mean_ds = mean_ds.merge(nuc)
mean_ds
```

```python
mean_ds.to_csv("./data/mean_dataset.csv", index=False)
```

#### Median

```python
compressed_df = df.groupby(["gene_id", "transcript_id", "position"]).median().reset_index()
compressed_df
```

```python
temp = df[["gene_id", "transcript_id", "position", "nucleotide"]].groupby(["gene_id", "transcript_id", "position"])["nucleotide"].unique().reset_index()
temp.nucleotide = temp.nucleotide.apply(lambda x: x[0])
temp
```

```python
compressed_df = compressed_df.merge(temp)
compressed_df
```

```python
compressed_df.to_csv("./data/median_dataset.csv", index=False)
```

#### Min-Max

```python
min_dataset = df.groupby(['gene_id', 'transcript_id', 'position']).min().reset_index()
max_dataset = df.groupby(['gene_id', 'transcript_id', 'position']).max().reset_index()
```

```python
max_dataset = max_dataset.rename(columns = {"dwell_1": "dwell_1_max",
                    "std_1": "std_1_max",
                    "mean_1": "mean_1_max",
                    "dwell_2": "dwell_2_max",
                    "std_2": "std_2_max",
                    "mean_2": "mean_2_max",
                    "dwell_3": "dwell_3_max",
                    "std_3": "std_3_max",
                    "mean_3": "mean_3_max"})
```

```python
minmax_data = pd.merge(min_dataset,max_dataset,on=['gene_id', 'transcript_id', 'position', "nucleotide", 'label'] ,how='left')
```

```python
column_to_move = minmax_data.pop("label")
minmax_data.insert(22, "label", column_to_move)
```

```python
minmax_data
```

```python
minmax_data.to_csv('./data/minmax_dataset.csv', header= True, index=False)
```
